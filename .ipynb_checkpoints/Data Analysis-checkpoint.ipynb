```r
gettimedtrans <- function(data, calcmean=F) {
	prevAOI <- ""
	totalfixdur <- 0
	transListDur <- NULL
	transListCnt <- NULL
	for (i in 1:length(data[,1])) {
		if (prevAOI != "") {
			if (prevAOI == data$AOIName[i]) {
				totalfixdur <- totalfixdur + data$FixationDurationInS[i]				
			} else {
				translabel <- paste(prevAOI, "->", data$AOIName[i], sep="")
				if (translabel %in% names(transListDur)) {
					transListCnt[[translabel]] <- transListCnt[[translabel]] + 1
					transListDur[[translabel]] <- transListDur[[translabel]] + totalfixdur
				} else {
					transListDur <- append(transListDur, totalfixdur)
					names(transListDur)[length(transListDur)] <- translabel
					transListCnt <- append(transListCnt, 1)
					names(transListCnt)[length(transListCnt)] <- translabel
					totalfixdur <- 0
				}
			}
		} 
		prevAOI <- data$AOIName[i]
	}
	
	# this returns the total fixation time per transition type
    
	if (calcmean) {
        
        transListDur/transListCnt 
    
    }else {
        
       transListDur

    }
    

	# this returns the average fixation time per transition type
	# transListDur/transListCnt # uncomment this line
}

# example use:
# mel_trans_fixdur <- gettimedtrans(data_mel_123) # get total fixation time per transition type
# mel_trans_fixdur
# mel_trans_fixave <- gettimedtrans(data_mel_123, calcmean=T) # get average fixation time per transition type
# mel_trans_fixave
```

## Answer Accuracy and Confidence Analysis
### Methodology 
Analyzing scenario answer required that we use a scoring mechanism. For each question, a participant will score either 0 or 1 for accuracy. ***Accurate*** answer receive a ***1*** and ***innacurate*** answers receive a ***0*** . The confidence level in the answer will be based on the manual analysis of the justification the participant gave for his answer. This analysis will be based on keywords used for answer justification. ***NO CONF***  will receive a score of ***0***, ***LOW CONF*** will receive a score of ***0.5*** and HIGH CONF will receive a score of ***1***.  
Situation where a participant the answer wrong with high confidence need to be investigated to understand the reason of this confusion

## System Usability Scale Scoring
### Overview of the method used in finding your SUS score.
The following ten questions about the UI were asked to each participants at the end of each session:
1. I think that I would like to use this system frequently.
2. I found the system unnecessarily complex.
3. I thought the system was easy to use.
4. I think that I would need the support of a technical person to be able to use this system.
5. I found the various functions in this system were well integrated.
6. I thought there was too much inconsistency in this system.
7. I would imagine that most people would learn to use this system very quickly.
8. I found the system very cumbersome to use.
9. I felt very confident using the system.
10. I needed to learn a lot of things before I could get going with this system.

Users have ranked each of the 10 questions above from 1 to 5, based on their level of agreement (Strongly disagree...Strongly agree).

 * For each of the odd numbered questions, we subtract 1 from the score.
 * For each of the even numbered questions, subtract their value from 5.
 
We take these new values which you have found, and add up the total score. Then multiply this by 2.5.
The result of all these tricky calculations is that you now have your score out of 100. This is NOT a percentage, but it is a clear way of seeing your score.
